```python
def lin(x): 
    return x

def relu(x):
    return np.maximum(x,0)

def heaviside(x):
    return np.array(x > 0)

def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-5,5, 1000)
y1 = np.tanh(x)
y2 = relu(x)
y3 = heaviside(x)
y4 = sigmoid(x)
```

```python
# import matplotlib as mpl
# mpl.rcParams['text.usetex'] = True

# mpl.rcParams['text.latex.preamble'] = r'\usepackage{{amsmath}}'

fig, ax = plt.subplots(2,2)

sns.lineplot(x, y1, ax = ax[0,0], palette = "tab10", color = "b")
sns.lineplot(x, y2, ax = ax[0,1], palette = "tab10", color = "b")
sns.lineplot(x, y3, ax = ax[1,0], palette = "tab10", color = "b")
sns.lineplot(x, y4, ax = ax[1,1], palette = "tab10", color = "b")
ax[0,0].set_title('Tanh')
ax[0,1].set_title('ReLu')
ax[1,0].set_title('Degrau')
ax[1,1].set_title('Sigmoid')

# ax[0,0].text(-3, 0.3, r'$\sigma(x) = x$')
# ax[0,1].text(-3, 2.5, r'$\sigma(x) = max(0,x)$')
# ax[1,0].text(-3, 0.3, r'$\begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$')
# ax[1,0].text(-3, 0.3, r'$\sigma(x) = \left\{\begin{array}{l} 1 , \text{ se x > 0} \\0, \text{, caso contrário}   \end{array}\right.$')
# ax[1,1].text(-3.5, 0.5, r'$\sigma(x) = \frac{1}{1+e^{-x}}$')

plt.tight_layout()

plt.show()
fig.savefig('/home/vitor/RepositorioGit/funcoes_de_ativacao.png')
```
